## PPM generator

```python
# Image dimensions
image_width = 256
image_height = 256

# Render
print("P3\n{} {}\n255".format(image_width, image_height))

for j in range(image_height):
    for i in range(image_width):
        r = i / (image_width - 1)
        g = j / (image_height - 1)
        b = 0

        ir = int(255.999 * r)
        ig = int(255.999 * g)
        ib = int(255.999 * b)

        print(ir, ig, ib)
```

### Using the progress indicator

```python
# Image dimensions
image_width = 256
image_height = 256

# Render
print("P3\n{} {}\n255".format(image_width, image_height))

for j in range(image_height):
    for i in range(image_width):
        r = i / (image_width - 1)
        g = j / (image_height - 1)
        b = 0

        ir = int(255.999 * r)
        ig = int(255.999 * g)
        ib = int(255.999 * b)

        print(ir, ig, ib)

    # Progress indicator
    progress = (j + 1) / image_height * 100
    print("\rProgress: [{:<50}] {:.2f}%".format("=" * int(progress // 2), progress), end="", flush=True)

print("\nDone!")
```

### Separate the image data and progress indicator

```python
import sys

# Image dimensions
image_width = 256
image_height = 256

# Output file
output_file = "output.ppm"

# Render
with open(output_file, 'w') as f:
    f.write("P3\n{} {}\n255\n".format(image_width, image_height))

    for j in range(image_height):
        for i in range(image_width):
            r = i / (image_width - 1)
            g = j / (image_height - 1)
            b = 0

            ir = int(255.999 * r)
            ig = int(255.999 * g)
            ib = int(255.999 * b)

            f.write("{} {} {}\n".format(ir, ig, ib))

        # Progress indicator
        progress = (j + 1) / image_height * 100
        sys.stdout.write("\rProgress: [{:<50}] {:.2f}%".format("=" * int(progress // 2), progress))
        sys.stdout.flush()

sys.stdout.write("\nDone!\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# Image dimensions
image_width = 256
image_height = 256

# Render
image = np.zeros((image_height, image_width, 3))

for j in tqdm(range(image_height), desc='Rendering', unit='row'):
    for i in range(image_width):
        r = i / (image_width - 1)
        g = j / (image_height - 1)
        b = 0
        image[j, i] = [r, g, b]

plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import time

# Image dimensions
image_width = 256
image_height = 256

# Generating coordinate grids with a progress bar
with tqdm(total=image_height, desc='Generating Image', unit='row') as pbar:
    for _ in range(image_height):
        time.sleep(0.01)  # Simulating a time-consuming task
        pbar.update(1)

    x = np.linspace(0, 1, image_width).reshape(1, -1)
    y = np.linspace(0, 1, image_height).reshape(-1, 1)

    # Creating the image
    image = np.zeros((image_height, image_width, 3))
    image[:, :, 0] = x      # Red channel
    image[:, :, 1] = y      # Green channel
    image[:, :, 2] = 0      # Blue channel

# Display the image
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
import math


# Decorator for checking Vector3 types
# Used this piece of code :
# https://stackoverflow.com/questions/15299878/how-to-use-python-decorators-to-check-function-arguments
def accepts(*types):
    def check_accepts(f):
        assert len(types) == f.__code__.co_argcount

        def new_f(*args, **kwds):
            for (a, t) in zip(args, types):
                assert isinstance(a, t), \
                       "arg %r does not match %s" % (a, t)
            return f(*args, **kwds)
        new_f.__name__ = f.__name__
        return new_f
    return check_accepts


class Vector3(list):

    def __init__(self, e0, e1, e2):
        self.append(float(e0))
        self.append(float(e1))
        self.append(float(e2))
        # XYZ coordinates
        self.x = e0
        self.y = e1
        self.z = e2
        # RGB
        self.r = e0
        self.g = e1
        self.b = e2
        # Length
        self.length = math.sqrt(e0*e0 + e1*e1 + e2*e2)
        self.squared_length = (e0*e0 + e1*e1 + e2*e2)

    def __add__(self, other):
        return Vector3(self[0] + other[0],
                       self[1] + other[1],
                       self[2] + other[2])

    def __sub__(self, other):
        return Vector3(self[0] - other[0],
                       self[1] - other[1],
                       self[2] - other[2])

    def __mul__(self, other):
        return Vector3(self[0] * other,
                       self[1] * other,
                       self[2] * other)

    def mul(self, other):
        return Vector3(self[0] * other[0],
                       self[1] * other[1],
                       self[2] * other[2])

    def __truediv__(self, other):
        return Vector3(self[0] / other,
                       self[1] / other,
                       self[2] / other)

    def make_unit_vector(self):
        """
        Makes the current vector it's unit vector
        :return:
        """
        self[0] = self[0]/self.length
        self[1] = self[1]/self.length
        self[2] = self[2]/self.length


@accepts(Vector3, Vector3)
def dot(v1, v2):
    return v1[0]*v2[0] + v1[1]*v2[1] + v1[2]*v2[2]


@accepts(Vector3, Vector3)
def cross(v1, v2):
    return Vector3(v1[1]*v2[2] - v1[2]*v2[1],
                   v1[2]*v2[0] - v1[0]*v2[2],
                   v1[0]*v2[1] - v1[1]*v2[0])


@accepts(Vector3)
def unit_vector(vector):
    return Vector3(vector[0] / vector.length, vector[1] / vector.length, vector[2] / vector.length)
```

# Object orientation coding and build the vector class

```python
import numpy as np
import matplotlib.pyplot as plt

class Vec3:
    def __init__(self, e0, e1, e2):
        self.e = np.array([e0, e1, e2])

    def x(self):
        return self.e[0]

    def y(self):
        return self.e[1]

    def z(self):
        return self.e[2]

    def __neg__(self):
        return Vec3(-self.e[0], -self.e[1], -self.e[2])

    def __getitem__(self, index):
        return self.e[index]

    def __add__(self, other):
        return Vec3(self.e[0] + other.e[0], self.e[1] + other.e[1], self.e[2] + other.e[2])

    def __sub__(self, other):
        return Vec3(self.e[0] - other.e[0], self.e[1] - other.e[1], self.e[2] - other.e[2])

    def __mul__(self, t):
        return Vec3(self.e[0] * t, self.e[1] * t, self.e[2] * t)

    def __rmul__(self, t):
        return self.__mul__(t)

    def __truediv__(self, t):
        return Vec3(self.e[0] / t, self.e[1] / t, self.e[2] / t)

    def length(self):
        return np.sqrt(np.sum(self.e ** 2))

    def to_np(self):
        return self.e

    def unit_vector(self):
        return self / self.length()


def write_color(pixel_color):
    return [int(255.999 * pixel_color.x()), int(255.999 * pixel_color.y()), int(255.999 * pixel_color.z())]

# Alias for color
Color = Vec3
```

```python
# Image dimensions
image_width = 256
image_height = 256

# Create image
image = np.zeros((image_height, image_width, 3), dtype=np.uint8)

# Render and set progress output
for j in range(image_height):
    for i in range(image_width):
        pixel_color = Color(i / (image_width - 1), j / (image_height - 1), 0)
        image[j, i] = write_color(pixel_color)

# Display the image
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
class Ray:
    def __init__(self, origin, direction):
        self.origin = Vec3(*origin)
        self.direction = Vec3(*direction)

    def at(self, t):
        return self.origin + t * self.direction
```

```python
def ray_color(r):
    unit_direction = r.direction.unit_vector()
    a = 0.5 * (unit_direction.y + 1.0)
    return (1.0 - a) * Vec3(1.0, 1.0, 1.0) + a * Vec3(0.5, 0.7, 1.0)
```

```python
import matplotlib.pyplot as plt

# Set up the image dimensions and aspect ratio
aspect_ratio = 16.0 / 9.0
image_width = 400
image_height = max(1, int(image_width / aspect_ratio))

# Camera settings
focal_length = 1.0
viewport_height = 2.0
viewport_width = viewport_height * (image_width / image_height)
camera_center = [0, 0, 0]

# Compute the dimensions of the image plane in world coordinates
# Compute the dimensions of the image plane in world coordinates
viewport_u = Vec3(viewport_width, 0, 0)
viewport_v = Vec3(0, -viewport_height, 0)
pixel_delta_u = viewport_u / image_width
pixel_delta_v = viewport_v / image_height
viewport_upper_left = Vec3(*camera_center) - Vec3(0, 0, focal_length) - viewport_u / 2 - viewport_v / 2
pixel00_loc = viewport_upper_left + 0.5 * (pixel_delta_u + pixel_delta_v)


# Create the image
image = np.zeros((image_height, image_width, 3))

for j in range(image_height):
    for i in range(image_width):
        pixel_center = pixel00_loc + i * Vec3(*pixel_delta_u) + j * Vec3(*pixel_delta_v)
        ray_direction = pixel_center - Vec3(*camera_center)
        r = Ray(camera_center, ray_direction.e)
        pixel_color = ray_color(r)
        image[j, i] = pixel_color.e

# Display the image
plt.imshow(image)
plt.axis('off')
plt.show()
```
